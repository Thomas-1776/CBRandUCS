```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv(list(uploaded.keys())[0])

# -------- INPUT & OUTPUT --------
X = df[['Fly ash','Lime','Cement','L.L','P.L','OMC','MDD']].values
y = df[['CBR','UCS']].values

# -------- NORMALIZATION --------
from sklearn.preprocessing import MinMaxScaler

scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

# -------- TRAIN-TEST SPLIT --------
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# -------- BUILD ANN MODEL --------
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()  #creates linear of stack layers

# Hidden layer (10 neurons)
model.add(Dense(10, input_dim=7, activation='tanh'))

# Output layer (CBR & UCS)
model.add(Dense(2, activation='tanh'))

# Compile (Levenberg-Marquardt not available in Keras в†’ using Adam)
model.compile(optimizer='adam', loss='mse')

# -------- TRAIN --------
model.fit(X_train, y_train, epochs=200, verbose=1)

# -------- PREDICT --------
y_pred_scaled = model.predict(X_test)  #Feeds the unseen test data into your trained network
#to see what scaled values it predicts for CBR and UCS.

# Convert back to original values (scaled data converted to original engineerin value)
y_pred = scaler_y.inverse_transform(y_pred_scaled)
y_actual = scaler_y.inverse_transform(y_test)

# -------- DISPLAY RESULTS --------
results = pd.DataFrame({
    'Actual_CBR': y_actual[:,0],
    'Predicted_CBR': y_pred[:,0],
    'Actual_UCS': y_actual[:,1],
    'Predicted_UCS': y_pred[:,1]
})

print(results.head(10))
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np # Ensure numpy is imported for sqrt

print("CBR R2:", r2_score(y_actual[:,0], y_pred[:,0]))
print("UCS R2:", r2_score(y_actual[:,1], y_pred[:,1]))

print("CBR RMSE:", np.sqrt(mean_squared_error(y_actual[:,0], y_pred[:,0])))
print("UCS RMSE:", np.sqrt(mean_squared_error(y_actual[:,1], y_pred[:,1])))
```
